```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# MST analysis (Figure 5e)

```python
dfs = pd.read_excel("data/231212_RBM12_MST_2.xlsx", sheet_name=None)
d1 = dfs["RBM12 MST"]
d2 = dfs["Fit-Values"]

m1 = pd.DataFrame(d1.values[1:,1:4])
m2 = pd.DataFrame(d1.values[1:,4:7])
means1, stds1 = m1.mean(axis=1), m1.std(axis=1)
means2, stds2 = m2.mean(axis=1), m2.std(axis=1)
xs = d1.values[1:,0].astype(float)
plt.figure(figsize=(4, 3))
plt.errorbar(xs, means1, yerr=stds1, fmt=".", lw=1, capsize=3, capthick=1, color="C2")
plt.errorbar(xs, means2, yerr=stds2, fmt=".", lw=1, capsize=3, capthick=1, color="C3")
if True:
    xs = d2["Dose"]
    ys1 = d2[["DeltaFNorm", "DeltaFNorm.1", "DeltaFNorm.2"]].mean(axis=1)
    ys2 = d2[["DeltaFNorm.3", "DeltaFNorm.4", "DeltaFNorm.5"]].mean(axis=1)
    plt.plot(xs, ys1, color="C2", label="RBM12-BCL-2 rG4")
    plt.plot(xs, ys2, color="C3", label="RBM12-BCL-2 mut")
plt.gca().set_xscale("log")
plt.xlabel("Ligand Concentration [M]")
plt.ylabel("$\Delta$ Fnorm [â€°]")
plt.xlim(1e-9, 1e-4)
plt.ylim(-5, 40)
for loc in ["top", "right"]:
    plt.gca().spines[loc].set_visible(False)
plt.legend(handlelength=1, frameon=False, loc="upper left")
plt.tight_layout()
plt.savefig("figures/20231212_RBM12_MST.pdf")
```
