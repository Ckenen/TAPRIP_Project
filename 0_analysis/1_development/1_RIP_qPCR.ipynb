```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
```

# Relative enrichment of XIST (Figure 1d)

```python
np.random.seed(0)

dat = pd.read_excel("data/20240119_JA1_RIP_qPCR_XIST_MALAT1.xlsx", sheet_name="Relative enrichment of XIST", index_col=0)
means = dat.mean(axis=1)
stds = dat.std(axis=1)
xs = np.arange(len(means))
w = 0.4
ylim = 15
colors = ["lightgrey", "skyblue", "skyblue", "skyblue"]
plt.figure(figsize=(2.6, 2.8))
plt.bar(xs, means, yerr=stds, error_kw=dict(capsize=4, lw=1), width=w, edgecolor="black", color=colors)
if True:
    w1 = w * 0.4
    for i in np.arange(len(dat)):
        xs1 = i + (np.random.rand(len(dat.columns)) - 0.5) * w1
        ys1 = dat.values[i]
        plt.scatter(xs1, ys1, s=20, zorder=10, edgecolor="black", color="none")
if True:
    for i in range(1, 4):
        vs1 = dat.values[0]
        vs2 = dat.values[i]
        res = ttest_ind(vs1, vs2)
        print("P-value:", res)
        if False:
            w2 = w/2
            x1, x2 = 0 - w2, i + w2
            x3 = (x1 + x2) / 2
            y = 10 + (i - 1) * 1.8
            plt.plot([x1, x2], [y, y], lw=1, color="black")
            plt.text(x3, y + 0.2, "%.1e" % res[1], ha="center", va="bottom")
plt.xticks(xs, dat.index, rotation=45)
plt.yticks([0, 3, 6, 9, 12, 15])
plt.ylabel(dat.index.name)
plt.xlim(min(xs) - 0.5, max(xs) + 0.5)
plt.ylim(0, ylim)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.tight_layout()
plt.savefig("figures/relative_enrichment_of_xist.pdf")
plt.show()
plt.close()
```
